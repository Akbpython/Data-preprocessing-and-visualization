```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv("India GDP 1960-2022.csv",delimiter=',')
data
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data.drop(["Unnamed: 0"],axis=1,inplace=True)
data
```

```python
data.rename(columns={'India GDP - Historical Data':'Year','India GDP - Historical Data.1':'GDP(B,$)','India GDP - Historical Data.2':'Per Capita($)','India GDP - Historical Data.3':'Growth(%)'},inplace=True)
data
```

```python
data.head()
```

```python
data.drop(0,axis=0,inplace=True)
data
```

```python
data.info()
```

```python
data.dtypes
```

```python
data.head()
```

```python
data['GDP(B,$)'] = data['GDP(B,$)'].replace('[\$\,\.\"B"]', '', regex=True).astype(int)
data['GDP(B,$)'] = data['GDP(B,$)']/100
data
```

```python
data['Per Capita($)']=data['Per Capita($)'].replace('[\$\,]','',regex=True).astype(float)
data
```

```python
data['Growth(%)']=data['Growth(%)'].replace(['%'],['0%'],regex=True)
data
```

```python
data['Growth(%)'] = data['Growth(%)'].replace('[\"%"\.]', '', regex=True).astype(int)
data['Growth(%)'] = data['Growth(%)']/100
data
```

```python
import matplotlib.pyplot as plt
from matplotlib import style
from matplotlib import figure
style.use('fivethirtyeight')
```

```python
plt.bar(data['Year'],data['GDP(B,$)'],color='Red')
plt.title('Year vs GDP')
plt.xlabel('Year')
plt.ylabel('GDP')
plt.gcf().set_size_inches(31,33)
plt.show()
```

```python
plt.bar(data['Year'],data['Per Capita($)'],color='Blue')
plt.title('Year vs Capita')
plt.xlabel('Year')
plt.ylabel('Capita')
plt.gcf().set_size_inches(39,19)
plt.show()
```

```python
plt.barh(data['Year'],data['Per Capita($)'],color='Blue')
plt.title('Year vs Capita')
plt.xlabel('Year')
plt.ylabel('Capita')
plt.gcf().set_size_inches(11,19)
plt.show()
```

```python
plt.bar(data['Year'],data['Per Capita($)'],color=(data['Per Capita($)']>1000).map({True:'Green',False:'Red'}))
plt.title('Year vs Capita')
plt.xlabel('Year')
plt.ylabel('Capita')
plt.gcf().set_size_inches(39,19)
plt.show()
```
